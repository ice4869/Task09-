```python
import requests #匯入 requests 套件，用於向網站發送請求
from bs4 import BeautifulSoup #匯入 BeautifulSoup 套件，用於解析網頁網頁HTML

response = requests.get('https://udn.com/news/breaknews/1') #使用requests發送GET請求，取得聯合新聞網首頁的 HTML 頁面
soup = BeautifulSoup(response.text) #使用BeautifulSoup 解析 HTML 頁面頁面 ， 產生 Beautiful Soup 物件 soup

news = [] #建立一個空的列表，用於存放新聞內容

for link in soup.find_all('h3' , class_='rounded-thumb__title')[:4]:
  #對首頁上的每一篇新聞，取得其網址並訪問其網頁
  news_url = link.a['href']
  news_response = requests.get('https://udn.com' + news_url )
  news_soup = BeautifulSoup(news_response.text)

  #從新聞頁面上取得新聞內容的 div 元素，並將其文本取出
  news_content = news_soup.find('div' , class_= 'article-content__paragraph').text.strip().replace('\n' , '')

  #將新聞內容加入到 news 列表中
  news.append(news_content)

#print(news) #印出所有新聞內容  





# 安裝 jieba 中文斷詞套件
!pip install jieba 

# 下載官方字典檔
!wget https://raw.githubusercontent.com/fxsjy/jieba/master/extra_dict/dict.txt.big


# 載入套件與字典檔
import jieba
jieba.load_userdict("dict.txt.big")





import jieba

#將每篇新聞文章斷詞後的結果存入 tokens 變數，並計算詞頻
tokens = []
word_count = {}
for d in news:
  #精確模式斷詞
  token = list(jieba.cut(d , HMM = False))
  tokens.append(token)
  #計算詞頻
  for w in token:
    if w not in word_count:
      word_count[w] = 1
    else:
      word_count[w] = word_count[w] + 1  

print(word_count)      
```
